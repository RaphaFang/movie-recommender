```python
import pandas as pd

ratings = pd.read_csv("row_data/MovieLens 20M Dataset/rating.csv")
movies = pd.read_csv("row_data/MovieLens 20M Dataset/movie.csv")

ratings_movies = pd.merge(ratings, movies, on="movieId", how="inner")

metadata = pd.read_csv("row_data/The Movies Dataset/movies_metadata.csv", low_memory=False)
```

```python
# 1. 多少評論者？多少電影？總共多少評論？
n_users = ratings["userId"].nunique()
n_movies = ratings["movieId"].nunique()
n_ratings = len(ratings)

print(n_users,n_movies,n_ratings)
```

```python
# 2.1 這邊要作的事情是處理整體的平紛紛不

rating_dist = ratings["rating"].value_counts().sort_index()
print(rating_dist)
```

```python
# 2.2 這邊是按照 genres 頻分數量分類
df_genre = ratings_movies.copy()
df_genre["genre"] = df_genre["genres"].str.split("|")
df_genre = df_genre.explode("genre")

df_without_blank = df_genre[(df_genre["genre"].notna()) & (df_genre["genre"] != "(no genres listed)")]

# r = df_without_blank.groupby('genre').agg(
#   mean = ("rating", "mean"),
#   count = ("rating","count")
# ).sort_values('mean', ascending=False)
r = df_without_blank['genre'].value_counts().reset_index()
print(r)
```

```python
# 2.3 find out the most viewed movie

rr = ratings_movies.groupby(["movieId", "title"]).size().reset_index(name="most_viewed_m").sort_values('most_viewed_m', ascending=False
).head(20)
rr

# top_movies = (
#     ratings_movies.groupby(["movieId", "title"])
#     .size()
#     .reset_index(name="rating_cnt")
#     .sort_values("rating_cnt", ascending=False)
#     .head(10)
# )
# ratings_movies
```
